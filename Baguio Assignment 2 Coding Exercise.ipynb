## A Bar chart(using matplotlib)

```python
tds_df = pd.read_csv("titanic_dataset.csv")
tds_df = tds_df.sort_values(by="PassengerId")

#calling two specific column to use for chart
tds_df1 = tds_df.values
x_1 = tds_df1[:,0]
y_1 = tds_df1[:,5]

#creating a figure by using plotlib
fig_1 = plt.figure(figsize = (8, 6))
axes_1 = fig_1.add_axes([2,2,2,1])

#setting a title for the bar chart
axes_1.set_title("Age range or each Passenger")

#since the ammount of passenger is huge lets just zoom the axes range from 0 to 200
axes_1.set_xlim([0,200])
axes_1.set_ylim([0,80])
axes_1.set_xlabel("PassengerId")
axes_1.set_ylabel("Age")
axes_1.bar(x_1, y_1, label = "Each bar line indicates one person\n"
           "Blank line gaps indicate of no age info")
axes_1.legend(loc=0)
```

# A wordcloud(using matplotlib)

```python
stopwords = set(STOPWORDS)

def Wordcloud(data, title = None):
    wordcloud = WordCloud(
        background_color = "White",
        stopwords = stopwords,
        max_words = 150,
        max_font_size = 40,
        scale = 3,
        ).generate(str(data))
    
    fig = plt.figure(1, figsize = (20, 20))
    plt.axis("off")
    if title:
        fig.suptitle(title, fontsize = 20)
        fig.subplots_adjust(top = 2.3)
    plt.imshow(wordcloud)
    
Wordcloud(tds_df["Name"].dropna())
```
